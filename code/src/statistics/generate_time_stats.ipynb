```python
from pathlib import Path
import json
```

```python
def load_results(results_dir: str) -> list[dict]:
    results = []
    for f in Path(results_dir).glob("*.json"):
        with open(f, "r", encoding="utf-8") as fh:
            results.append(json.load(fh))
    return results

def calculate_total_time(results):
    total = 0.0

    for r in results:
        total += r["timings"]["total_seconds"]

    avg = total / len(results) if results else 0.0
    print(f"Total time (s): {total:.2f}")
    print(f"Avg time (s)  : {avg:.2f}")
    return total, avg
```

```python
dataset = "2wiki"
mode = "psp"

if mode != "psp":
    ft_case = f"../counterfactuals/results/{dataset}/delete_ops_ft"
    ff_case = f"../counterfactuals/results/{dataset}/all_ops_ff"

    results = load_results(ft_case + ff_case)
    calculate_total_time(results)
else: 
    psp_case = f"/home/gbalanos/GloRAG-Ex/code/src/counterfactuals/results/psp/{dataset}/delete_ops_ft"

    results = load_results(psp_case)
    calculate_total_time(results)
```
