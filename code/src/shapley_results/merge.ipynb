```python
import json

with open("/home/gbalanos/GloRAG-Ex/competitors/Shapley/shapley_text_results/synthetic_shapley_text_sentence_tmc.json") as f:
    attribution = json.load(f)

with open("/home/gbalanos/GloRAG-Ex/code/benchmark/results/comparison_synthetic_2.json") as f:
    eval_results = json.load(f)

with open("/home/gbalanos/GloRAG-Ex/competitors/Shapley/shapley_text_results/synthetic_shapley_text_sentence_tmc_metrics.json") as f:
    metrics = json.load(f)

ft_cases = eval_results["cases"]["ft"]
ff_cases = eval_results["cases"]["ff"]

def build_merged(case_ids):
    merged = {}
    for idx in case_ids:
        idx_str = str(idx)
        result = eval_results["results"].get(idx_str, {})
        attr = attribution.get(idx_str, {})
        met = metrics.get(idx_str, {})

        merged[idx_str] = {
            "question": result.get("question"),
            "ground_truth": result.get("ground_truth"),
            "rag_answer": result.get("rag_answer"),
            "llm_answer": result.get("llm_answer"),
            "rag_score": result.get("rag_score"),
            "llm_score": result.get("llm_score"),
            # attribution fields
            "n_items": attr.get("n_items"),
            "shapley_scores": attr.get("shapley_scores", {}),
            # metrics fields
            "rag_calls": met.get("rag_calls"),
            "rag_time": met.get("rag_time"),
            "shap_utility_evals": met.get("shap_utility_evals"),
            "shap_forward_passes": met.get("shap_forward_passes"),
            "shap_time": met.get("shap_time"),
            "judge_calls": met.get("judge_calls"),
            "judge_time": met.get("judge_time"),
            "total_calls": met.get("total_calls"),
            "total_time": met.get("total_time"),
        }
    return merged

ft_merged = build_merged(ft_cases)
ff_merged = build_merged(ff_cases)

with open("/home/gbalanos/GloRAG-Ex/competitors/Shapley/shapley_text_results/synthetic/shap_ft.json", "w") as f:
    json.dump(ft_merged, f, indent=2)

with open("/home/gbalanos/GloRAG-Ex/competitors/Shapley/shapley_text_results/synthetic/shap_ff.json", "w") as f:
    json.dump(ff_merged, f, indent=2)

print(f"ft_merged.json: {len(ft_merged)} cases")
print(f"ff_merged.json: {len(ff_merged)} cases")
```
