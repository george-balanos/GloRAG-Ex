```python
import json
import sys

def summarize(path: str) -> None:
    with open(path) as f:
        data = json.load(f)

    total_calls = 0
    total_time = 0.0

    print(f"{'ID':<5} {'LLM Calls':>10} {'Time (s)':>10}")
    print("-" * 28)

    for key, item in data.items():
        secs = item["total_time"]
        calls = item["total_calls"]
        total_calls += calls
        total_time += secs
        print(f"{key:<5} {calls:>10} {secs:>10.2f}")

    print("-" * 28)
    print(f"{'TOTAL':<5} {total_calls:>10} {total_time:>10.2f}")
    print(f"\nQuestions : {len(data)}")
    print(f"LLM calls : {total_calls}")
    print(f"Total time: {total_time:.2f}s ({total_time / 60:.1f} min)")
    print(f"Avg calls : {total_calls / len(data):.1f} per question")
    print(f"Avg time  : {total_time / len(data):.2f}s per question")
```
