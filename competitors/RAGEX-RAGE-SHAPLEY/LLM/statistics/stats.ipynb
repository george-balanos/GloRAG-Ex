```python
import json
import sys
from collections import defaultdict


def print_stats(label: str, cases: list):
    llm_calls = [c["llm_calls"] for c in cases if "llm_calls" in c]
    times = [c["elapsed_time"] for c in cases if "elapsed_time" in c]
    print(f"  Cases found:          {len(cases)}")
    print(f"  Avg LLM calls:        {sum(llm_calls) / len(llm_calls):.2f}" if llm_calls else "  Avg LLM calls:        N/A")
    print(f"  Avg time (seconds):   {sum(times) / len(times):.2f}" if times else "  Avg time (seconds):   N/A")
    print(f"  Total time (seconds): {sum(times):.2f}" if times else "  Total time (seconds): N/A")


def compute_stats(filepath: str):
    with open(filepath) as f:
        data = json.load(f)

    cases = data.get("cases", [])
    if not cases:
        print("No cases found.")
        return

    groups = defaultdict(list)
    for c in cases:
        groups[c.get("case_type", "unknown")].append(c)

    print("=== Overall ===")
    print_stats("all", cases)

    for case_type, group in sorted(groups.items()):
        print(f"\n=== case_type: '{case_type}' ===")
        print_stats(case_type, group)
```
